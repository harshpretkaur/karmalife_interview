```python
import pandas as pd ##importing necessary lirbaries
```

```python
# # Question 1: Some of the orders are stored in another csv file named `bigkart_newsales`.
# Read the csv file, store it in a data frame and add it to the `bigkart_sales` data frame. 
# Find the total sales value of the category 'Office Supplies' after combining the dataframes
# # #
```

```python
##reading files into two separate dataframes
df1=pd.read_csv("bigkart_sales.csv")
df2=pd.read_csv("bigkart_newsales.csv")
```

```python
##concatenating the dataframes
df=df1.append(df2,ignore_index=True)
```

```python
df ##result
```

```python
ans_one=df.groupby('Category').agg({'Sales':sum})
##grouping by category
```

```python
ans_one
```

```python
"""Total sales for Office Supplies: 7970"""
```

### Question 4

```python
# Question 4: How many invalid order IDs are there in the data frame. 
# An order id is of the form AZ-2011-Y where Y represents a whole number. 
# A Order ID is said to be valid only if Y consists of 7 digits. 
# Find the number of invalid order order IDs in the data frame.
```

```python
##splitting into a list
dfnew['check']=df['Order ID'].apply(lambda x: x.split("-"))
```

```python
dfnew
```

```python
dfnew["check"]=dfnew['check'].apply(lambda x: (x[2]))
###preserving only the last item of the list to check for no. of digits
```

```python
##column type conversion
dfnew.astype(str)
```

```python
dfnew["check"]=dfnew['check'].apply(lambda x: len(x))
##computing the length
```

```python
dfnew
```

```python
ans_four=dfnew.groupby(['check']).size()
```

```python
ans_four
```

```python
"""number of invalid ids:7"""
```
